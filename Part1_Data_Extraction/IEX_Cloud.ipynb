```python
import pandas as pd
import json
import numpy as np
import requests
```

```python
API_key = "pk_4857117b5f704dffb567856c1b3886a6"
symbol = "msft"
timeframe = "2m"
data_site = f'https://api.iex.cloud/v1/data/core/historical_prices/{symbol}?range={timeframe}&token={API_key}'

price_data = requests.get(data_site)
json_data = price_data.json()
```

```python
data_list = []
for item in json_data:
    data = {
        'Date': item['priceDate'],
        'Ticker': item['symbol'],
        'Open': item['open'],
        'High': item['high'],
        'Low': item['low'],
        'Close': item['close'],
        'Adj Close': item['fclose'],
        'Volume': item['volume']
    }
    data_list.append(data)

# Creating DataFrame
df = pd.DataFrame(data_list)

# Displaying DataFrame
df
```
